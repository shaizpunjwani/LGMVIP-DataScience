**Image to Pencil Sketch with Python:**


```python

# these are some usefull libraries to perform this task!
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

from skimage import io
img = io.imread('sample_data/cat.jpg')
Realimage=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(Realimage)
plt.axis(False)
plt.show()
```

```python
from google.colab.patches import cv2_imshow
gray_img=cv2.cvtColor(Realimage,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
#converted into grayscaled
```

```python
invert_image=cv2.bitwise_not(gray_img)
cv2_imshow(invert_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
#inverted image
```

```python
blur_img=cv2.GaussianBlur(invert_image,(7,7),0)
cv2_imshow(blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
#blurred the image
```

```python
invert_blur_img=cv2.bitwise_not(blur_img)
cv2_imshow(invert_blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

final_sketch=cv2.divide(gray_img,invert_blur_img,scale=256.0)
cv2.imwrite("convert_img.png",final_sketch)
cv2_imshow(final_sketch)
cv2.waitKey(0)
cv2.destroyAllWindows()
#converted into our pencil sketch
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Original', size=18)
plt.imshow(Realimage)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketched', size=18)
rgb_sketch=cv2.cvtColor(final_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
